```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
f = h5py.File('solution.h5', 'r')
dset = f['sol_vec']
transfer = np.zeros(dset.shape)
dset.read_direct(transfer, source_sel=np.s_[:,:], dest_sel=np.s_[:,:])
print(dset.shape)
f.close()
```

```python
petsc_data = transfer[:-1,:,0]+1j*transfer[:-1,:,1]
petsc_data = petsc_data.transpose()
del transfer
```

```python
plt.imshow(np.abs(petsc_data),cmap='jet',origin='lower')
plt.show()
```
