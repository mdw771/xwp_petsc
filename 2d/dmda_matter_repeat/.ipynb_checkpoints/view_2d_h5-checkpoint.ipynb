```python
dset = f['sol_vec']
print(dset.shape)
dim_x,dim_y = int(dset.shape[0]),int(dset.shape[1])
slices = 1
```

```python
data1 = np.zeros((dim_x,dim_y))
data2 = np.zeros((dim_x,dim_y))
```

```python
dset.read_direct(data1, source_sel=np.s_[:,:,0], dest_sel=np.s_[:,:])
dset.read_direct(data2, source_sel=np.s_[:,:,1], dest_sel=np.s_[:,:])
f.close()
```

```python
from multislice.prop import prop1FT

step_xy = 1.0987669393246516e-09
energy = 12000                     
wavel = (1239.84/energy)*10**(-9)  
step_z = 0.0001
L = step_xy*dim_x
wave_focus,L2 = prop1FT(test,step_xy,L,wavel,step_z)
#wave_focus,L2 = prop1FT(test - np.ones(np.shape(test)),step_xy,L,wavel,step_z)
#wave_focus = wave_focus +  np.ones(np.shape(test))
```

```python
from multislice import prop_utils
focal_spot_size = 50
focal_spot,x_,y_,max_val = prop_utils.get_focal_spot(np.abs(wave_focus),dim_x,focal_spot_size)
```

```python
plt.imshow(focal_spot,extent = np.array([-1,1,-1,1])*(focal_spot_size)*(L2/dim_x)*(1e9),cmap='jet')
plt.title('focal spot intensity (log scale)')
plt.xlabel('axes in nm')
plt.colorbar()
plt.show()
```

```python
def get_primary_lobe(foc_spot,plot=0):
    cx,cy = np.round(np.where(foc_spot==np.max(foc_spot)))
    nr,nc = np.shape(foc_spot)
    r = np.arange(nr)-cx 
    c = np.arange(nc)-cy 
    [R,C] = np.meshgrid(r,c)
    index = np.round(np.sqrt(R**2+C**2))+1 
    primary_lobe = 0
    temp = np.max(foc_spot)

    for i in range(1,int(np.max(index))):
        if np.max(foc_spot[np.where(index==i)])<=temp:
            primary_lobe+=np.sum(foc_spot[np.where(index==i)]**2)
            temp=np.max(foc_spot[np.where(index==i)])
            j = i
        else :
            print(j)
            break
    
    if plot==1 :
        plt.imshow(foc_spot,cmap='jet')
        plt.colorbar()
        for i in range(1,j):
            R1,C1 = np.where(index==i)
            plt.scatter(C1,R1,s=5,c='k',alpha=0.25)
        plt.show()
    return primary_lobe
```

```python
get_primary_lobe(focal_spot,1)
```
